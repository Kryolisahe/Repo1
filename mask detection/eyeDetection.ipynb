```python
import numpy as np
import random
from IPython.display import Image
import cv2
```

```python
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eye_cascade=cv2.CascadeClassifier("haarcascade_eye.xml")
```

```python
white=(255,255,255)
red=(0,0,255)
blue=(255,0,0)
thickness=2
fontSize=0.6
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
img=cv2.imread("image1.jpg")
#convert to gray
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#detect face
faces=face_cascade.detectMultiScale(gray,1.1,4)
#if face detected
if(len(faces)>0):
    for(x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),white,thickness)
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=img[y:y+h,x:x+w]
        
        #detect eyes
        eye_boundary=eye_cascade.detectMultiScale(gray,1.5,5)
        
        #face detected but eyes not detect means person is wearing sunglasses
        if(len(eye_boundary)==0):
            cv2.putText(img,"Sunglasses Present",(x+60,y+15),font,fontSize,blue,thickness,cv2.LINE_AA)
        else:
            for(ex,ey,ew,eh) in eye_boundary:
                
                #face and eyes are detcted and eye coordinates are within face_coordinates means that there are no sunglasses present
                if(y<ey<y+h):
                    cv2.putText(img,"No Sunglasses",(ex,ey),font,fontSize,red,thickness,cv2.LINE_AA)
                    
cv2.imwrite("SunglassesDetection.jpg",img)
display(Image(filename=="SunglassesDetection.jpg"))
                
            
    
```
